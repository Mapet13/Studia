> 1.Wyświetl wykaz zamówień, zawierający: id zamówienia, cenę jednostkową, ilość, wartość, cenę jednostkową powiększoną o 15%, wartość, a także kwoty rabatu w dwóch wariantach (dla ceny oryginalnej oraz ceny zmienionej)

```sql
SELECT 
    OrderID,
    UnitPrice, 
    Quantity, 
    UnitPrice * (1.15) AS [Cena powiększona], 
    UnitPrice * (Discount) AS [Rabat dla oryginalnej ceny], 
    (UnitPrice * 1.15) * (Discount) AS [Rabat dla powiększonej ceny]
FROM 
    [Order Details]
```

> 2.Wyświetl wykaz pracowników, w układzie: "\[zwrot grzecznościowy\] \[imię\] \[nazwisko\], ur. \[data\], zatrudniony w dniu \[data\], adres: \[adres\], \[miasto\], \[kod\], \[państwo\]", posortowany rosnąco po dacie urodzenia.

```sql
SELECT 
    TitleOfCourtesy + ' ' + FirstName + ' ' + LastName + 
    ', ur. ' + CAST(CONVERT(date, BirthDate) AS varchar) + ', zatrudniony w dniu ' + CAST(CONVERT(date, HireDate) AS varchar)
    + ', adres: ' + Address + ', ' + City + ', ' + Country 
    AS [Wykaz Pracowników]
FROM 
    Employees
ORDER BY
    BirthDate
```
